```python
import pandas as pd
import numpy as np
import os

# 1. Dynamically read all CSV files from the folder
folder_path = "data/"
temp_list = []

# Loop through all files in the directory
for filename in os.listdir(folder_path):
    if filename.endswith(".csv"):
        file_full_path = os.path.join(folder_path, filename)
        
        data = pd.read_csv(file_full_path)
        
        # Clean column names (stripping whitespace)
        data.columns = [c.strip() for c in data.columns]
        temp_list.append(data)

# Combine everything
if temp_list:
    options_df = pd.concat(temp_list, axis=0, ignore_index=True)
else:
    raise FileNotFoundError("No CSV files found in the specified directory.")

# 2. Standardize Dates and Basic Math
options_df['Date'] = pd.to_datetime(options_df['Date'], dayfirst=True)
options_df['Expiry'] = pd.to_datetime(options_df['Expiry'], dayfirst=True)

# Time to Maturity (T) in years
options_df['T'] = (options_df['Expiry'] - options_df['Date']).dt.days / 365.0

# 3. Filter: Keep only tradable days (T > 0) and sort by Date
options_df = options_df[options_df['T'] > 0].copy()
options_df = options_df.sort_values(by='Date')

print(f"Total options records merged: {len(options_df)}")
print(f"Date range: {options_df['Date'].min().date()} to {options_df['Date'].max().date()}")
```

```python
options_df.to_csv("merged_options_data.csv", index=False)
```
